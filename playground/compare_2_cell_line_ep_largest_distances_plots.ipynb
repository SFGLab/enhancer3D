```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

gm12878_hffc6_denisovan_genes_with_biggest_distance_difference_df = pd.read_csv('./gm12878_hffc6_denisovan_genes_with_biggest_distance_difference_df.csv')
gm12878_hffc6_denisovan_genes_with_biggest_distance_difference_df
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x='gene_id', y='dist_diff_avg', data=gm12878_hffc6_denisovan_genes_with_biggest_distance_difference_df)
plt.xticks(
    rotation=90,
    fontsize=2,
    color='white'
)

plt.xlabel('Gene ID')
plt.ylabel('Distance Difference')

plt.title('Biggest Distance Difference between GM12878 and HFFC6 Neanderthal')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x='gene_id', y='dist_diff_avg', data=gm12878_hffc6_denisovan_genes_with_biggest_distance_difference_df[:100])
plt.xticks(
    rotation=90,
    fontsize=8,
)

plt.xlabel('Gene ID')
plt.ylabel('Distance Difference')

plt.title('Biggest Distance Difference between GM12878 and HFFC6 Neanderthal. Top 100')

plt.tight_layout()
plt.show()
```
